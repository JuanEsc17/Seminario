```python
rounds = [
 {
 'Shadow': {'kills': 2, 'assists': 1, 'deaths': True},
 'Blaze': {'kills': 1, 'assists': 0, 'deaths': False},
 'Viper': {'kills': 1, 'assists': 2, 'deaths': True},
 'Frost': {'kills': 0, 'assists': 1, 'deaths': False},
 'Reaper': {'kills': 1, 'assists': 1, 'deaths': False}
 },
 {
 'Shadow': {'kills': 0, 'assists': 2, 'deaths': False},
 'Blaze': {'kills': 2, 'assists': 0, 'deaths': True},
 'Viper': {'kills': 1, 'assists': 1, 'deaths': False},
 'Frost': {'kills': 2, 'assists': 1, 'deaths': True},
 'Reaper': {'kills': 0, 'assists': 1, 'deaths': False}
 },
 {
 'Shadow': {'kills': 1, 'assists': 0, 'deaths': False},
 'Blaze': {'kills': 2, 'assists': 2, 'deaths': True},
 'Viper': {'kills': 1, 'assists': 1, 'deaths': True},
 'Frost': {'kills': 0, 'assists': 1, 'deaths': False},
 'Reaper': {'kills': 1, 'assists': 1, 'deaths': False}
 },
 {
 'Shadow': {'kills': 2, 'assists': 1, 'deaths': False},
 'Blaze': {'kills': 1, 'assists': 0, 'deaths': True},
 'Viper': {'kills': 0, 'assists': 2, 'deaths': False},
 'Frost': {'kills': 1, 'assists': 1, 'deaths': True},
 'Reaper': {'kills': 1, 'assists': 1, 'deaths': False}
 },
 {
 'Shadow': {'kills': 1, 'assists': 2, 'deaths': True},
 'Blaze': {'kills': 0, 'assists': 1, 'deaths': False},
 'Viper': {'kills': 2, 'assists': 0, 'deaths': True},
 'Frost': {'kills': 1, 'assists': 1, 'deaths': False},
 'Reaper': {'kills': 1, 'assists': 1, 'deaths': True}
 }
]

stats = {
    'Shadow': {'kills': 0, 'assists': 0, 'deaths': 0},
    'Blaze': {'kills': 0, 'assists': 0, 'deaths': 0},
    'Viper': {'kills': 0, 'assists': 0, 'deaths': 0},
    'Frost': {'kills': 0, 'assists': 0, 'deaths': 0},
    'Reaper': {'kills': 0, 'assists': 0, 'deaths': 0}
}

mvps = {
    'Shadow': 0,
    'Blaze': 0,
    'Viper': 0,
    'Frost': 0,
    'Reaper': 0
}

scores = {
    'Shadow': 0,
    'Blaze': 0,
    'Viper': 0,
    'Frost': 0,
    'Reaper': 0
}

# Inicializa el índice de ronda
i = int(0)

score_round = {
    'Shadow': 0,
    'Blaze': 0,
    'Viper': 0,
    'Frost': 0,
    'Reaper': 0
}

# Se procesa cada ronda
for round in rounds:
    #Incremento el índice de ronda
    i += 1
    
    # Se procesan las estadísticas de cada jugador
    for player in round:
        # Se actualizan las estadísticas totales
        stats[player]['kills'] += round[player]['kills']
        stats[player]['assists'] += round[player]['assists']
        stats[player]['deaths'] += 1 if round[player]['deaths'] else 0

        # Se calculan los puntos de la ronda
        score_round[player] = (round[player]['kills'] * 3) + (round[player]['assists'] * 1) + (-1 if round[player]['deaths'] else 0)
        # Se calculan los puntos totales
        scores[player] += score_round[player]

    # Se verifica si el jugador es MVP
    mvp_player = max(score_round, key=lambda x: score_round[x])
    mvps[mvp_player] += 1

    # Ordeno los jugadores por puntaje
    sorted_players = sorted(scores.items(), key=lambda x: x[1], reverse=True)

    # Se imprimen los resultados de la ronda
    print(f"Ranking ronda: {i}")
    print(f"El MVP fue: {mvp_player}")
    print("Jugador    Kills    Asistencias  Muertes  Puntos")
    print("--------------------------------------------------")
    for player, x in sorted_players:
        # Se imprime el ranking de cada jugador
        print(f"{player:<10} {round[player]['kills']:<8} {round[player]['assists']:<12} {round[player]['deaths']:<8} {score_round[player]:<6}")
    print("--------------------------------------------------")

# Imprime el ranking final
print(f"Ranking final:")
print("Jugador    Kills    Asistencias  Muertes  MVPs   Puntos")
print("--------------------------------------------------------")
for player, x in sorted_players:
    # Se imprime el ranking de cada jugador
    print(f"{player:<10} {stats[player]['kills']:<8} {stats[player]['assists']:<12} {stats[player]['deaths']:<8} {mvps[player]:<6} {scores[player]:<6}") 
print("--------------------------------------------------------")   
```
